```python
x=np.linspace(-np.pi,np.pi,endpoint=True)
cos=np.cos(x)
sin=np.sin(x)
tan=np.tan(x)
plt.plot(x,cos,label="cos")
plt.plot(x,sin,label="sin")
plt.plot(x,tan,label="tan")
plt.ylim(-2,2)
plt.grid(color='g', linestyle='-', linewidth=0.5)
plt.legend(loc="upper right ")
plt.show()
```

```python
df = pd.DataFrame(np.random.randint(61,361,size=35).reshape((7,5))
                  , index=pd.Index(['Sun','Mon','Tue','Wed','Thu','Fri','Sat']
                  , name='Day')
                  , columns=pd.Index(['A','B','C','D','E']
                                     , name='Student'))
df.plot(kind='bar')
plt.title("Bar Chart of Students' Internet Usage")
plt.ylabel("Days of the Week")
plt.xlabel("Internet Usage (in minutes)")
plt.show()
```
